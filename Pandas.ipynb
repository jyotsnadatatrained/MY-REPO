```python
rto = pd.DataFrame({'Adhar':['A123','B123'],
                  'vehicle':['honda','yamaha'],
                  'license':['yes','no']})
rto
```

```python
final_merge = pd.merge(rto,insurance,on='Adhar')
final_merge
```

```python
claim_accepted = []
for claim in final_merge['license']:
    if claim == 'yes':
        claim_accepted.append('Accepted')
    else:
        claim_accepted.append('Rejected')
final_merge['claim_accepted ?'] = claim_accepted
final_merge
```

```python
#Creating dataframe with missing values

raw_data = {'name': ['rag',np.nan,'hds','fdgfh'],
           'age': [12,np.nan,23,54],
           'height': [32,np.nan,np.nan,np.nan],
           'sex': ['m',np.nan,np.nan,'f'],
           'unit1':[23,np.nan,12,34],
           'unit2':[56,np.nan,32,21]}
df = pd.DataFrame(raw_data)
df
```

```python
df.isnull().sum()
```

```python
#both isnull and isna are same

df.isna().sum()
```

```python
#notnull is opposite of isnull it shows how many rows do not have nulls

df.notnull().sum()
```

```python
#drop missing observation

df_no_missing = df.dropna()
df_no_missing
```

```python
df
```

```python
  #if we want to clean whole row with nan
    
df_cleaned = df.dropna(how='all')
df_cleaned
```

```python
#create a new column full of missing values

df['Final_score'] = np.nan
df
```

```python
df.dropna(axis=1, how = 'all')
```

```python
df
```

```python
#fiiling unit 1 with the mean value of unit 1

df['unit1'].fillna(df['unit1'].mean(), inplace = True)
df
```

```python
#filling unit1 with each sexs mean value of unit 2

df['unit2'].fillna(df.groupby('sex')['unit2'].transform('mean'), inplace = True)
df
```

```python
#backfill

df.fillna(method='bfill')
```

```python
df.fillna(method='ffill', inplace=True)
```
